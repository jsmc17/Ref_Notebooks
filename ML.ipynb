## Making Data-Driven Financial Decisions (NPV, IRR, EAA)

Common profitability analysis methods:
* Net Present Value (NPV)
* Internal Rate of Return (IRR)
* Equivalent Annual Annuity (EAA)

**Net Present Value** = sum of all discounted cash flows

Projects with higher cash flows and lower discount rates will have higher NPVs

NPV is simple measure that does not allow for comparison of projects of different sizes/lengths.

**Internal Rate of Return** = discount rate when NPV is set to 0

IRR can be used to compare projects of different sizes and lengths but does not allow comparison of total value of projects

```
p1 = np.array([-100, 150,200])
irr = npf.irr(p1)* 100
print(f'The IRR of the project is : {round(irr,2)}%')
```

```
npf.npv(rate = irr/100, values = [-100,150,200])
#basically 0
```

**Weighted Average Cost of Capital (WACC)** - the cost of funding a business (equity and debt) <br>
Consists of:
* Percent of company's financing from equity
* Percent of company's financing from debt
* Cost of company's equity (expected return from investors)
* Cost of company's debt (interest rate)
* Tax Rate 

```
percent_equity = .8
percent_debt = .2
cost_equity = .14   
cost_debt = .12
TR = .35

WACC = percent_equity * cost_equity + percent_debt * cost_debt * (1-TR)
WACC
```

Companies use WACC as discount rate because any project undertaken should outperform the cost of undertaking that action

WACC much higher than inflation - companies expect greater return 

Comparing two projects of different life spans

```
projects = pd.DataFrame({'Year': range(1,9), 'P1': [-100, 200,300,0,0,0,0,0], 'P2':[-125,100,100,100,100,100,100,100]}).set_index('Year')
projects
```

```
npv1 = npf.npv(.05, projects['P1'].values)
npv2 = npf.npv(.05, projects['P2'].values)
print(f'NPV of project 1: {npv1}\nNPV of project 2: {npv2}')
```

```
irr1 = npf.irr(projects['P1'].values)
irr2 = npf.irr(projects['P2'].values)
print(f'IRR of project 1: {irr1}\nIRR of project 2: {irr2}')
```

Project 2 is longer and has a higher NPV, however Project 1 has a higher IRR. Which is actually better?

**Equivalent Annual Annuity** (EAA) can be used to compare projects of different lifespans in present value terms

```
EAA1 = npf.pmt(rate = 0.05, nper = 3, pv = -1* npv1, fv = 0)
EAA2 = npf.pmt(rate = 0.05, nper = 8, pv = -1* npv2, fv = 0)
print(f'EAA of project 1: {EAA1}\nEAA of project 2: {EAA2}')
```

Project 1 is better despite having lower NPV (due mostly to short lifespan of projected cashflows)

### Window Functions

Windows functions in pandas
* Identify sub periods of your time series
* Calculate metrics for sub periods inside window
* Create a new time series of metrics
  * Each data point represents a summary of several data points of the original timseries
* Two types:
  * Rolling - same size, sliding
  * Expanding - growing, contains all prior values

#### Rolling Window

Calculating a rolling average

```
google = yf.Ticker("GOOGL").history(period="max").loc['2015':'2016']
google['Close'].rolling(window = 30).mean()
#window contains 30 previous observations/trading days
```

```
google['Close'].rolling(window = '30D').mean()
#window contains 30 previous calendar days
```

```
google['r90'] = google['Close'].rolling(window = 90).mean()
google[['Close', 'r90']].plot()
#new time series is much smoother because each point is average of last 90 days
```

# 3-Intro to Portfolio Analysis in Python

```
import pandas as pd
import numpy as np
```

Portfolio - Risk Management

Diversification - reduce fluctuations in value over short term, reduce overall risk

* Portfolio: a collection of investments (stocks,bonds,commodities,other) often owned by an individual
* Fund: a pool of investments that is managed by a professional fund manager. Individual investors buy "units" of the fund and the manager invests the money.
* Index: A smaller sample of the market that is representative of the whole (S&P500, Nasdaq, Russell2000, MSCI World Index

Active vs Passive investing:
* Passive investing: following a benchmark as closely as possible
* Active investing: taking active "betss" that are different from a benchmark

Portfolio Strategies:
* Equal Weighted
* Market Cap Weighted

Portfolio weights - % composition of a particular asset in a portfolio

Portfolio returns

* Changes in value over time
* Historic average returns often used to calculate expected return 
* Different types of returns
  * average return (geometric mean of a return series over a given time span)
  * cumulative return (total return over a period)
  * active return (relative performnace to a benchmark)
  * annualized return

Measuring risk of a portfolio:

* Investing is risky: individual assets will go up or down
* Expected return is a random variable
* Returns spread around the mean is measured by variance - a common measure of volatility

Correlation - increases risk of portfolio as a whole

>**Portfolio Variance** takes into account the individual assets' variances, the weights of the assets in the portfolio, as well as their correlation to each other

```
cov_matrix_d = returns.iloc[:,0:3].cov()*251  # multiply by 251 - number of trading days
cov_matrix_d
#TSLA has highest individual variance
```
