## Operations

Addition, multiplication, summing over dimensions

**Basic Operations**

TensorFlow Operation: 

image.png

TensorFlow graph contains edges and nodes, where the edges are tensors and the nodes are operations

```
from tensorflow import constant, add

# Define 0-dimensional tensors
A0 = constant([1])
B0 = constant([2])

# Define 1-dimensional tensors
A1 = constant([1,2])
B1 = constant([3,4])

# Define 2-dimensional tensors
A2 = constant([[1,2], [3,4]])
B2 = constant([[5,6], [7,8]])
```

Applying the **addition** operator

```
# scalar adition 
c0 = add(A0, B0)
c0
```

```
# vector addition
c1 = add(A1, B1)
c1
```

```
# matrix addition
c2 = add(A2, B2)
c2
```

* The `add()` operation performs **element-wise addition** with two tensors
* Each pair of tensors must have the same shape


Applying the **multiplication** operator

* Element-wise multiplication with `multiply()` operation
  * Tensors involved must have the same shape 
* Matirx multiplication performed with the `matmul()` operator
  * Number of columns of A must equal the number of rows of B 

```
from tensorflow import ones, matmul, multiply

A0 = ones(1)
A31 = ones([3,1])
A34 = ones([3,4])
A43 = ones([4,3])
```

**Summing over tensor dimensions**

* `reduce_sum()` operator sums over the dimensions of a tensor
  * reduce_sum(A) - sums over all dimensions of A
  * reduce_sum(A, i) - sums over dimension i

```
from tensorflow import ones, reduce_sum

# Define a 2x3x4 tensor of ones
A = ones([2,3,4])
A
```

```
# Sum over all dimensions
reduce_sum(A)
```

```
# Sum over dimensions of 0, 1, 2 
B0 = reduce_sum(A, 0)
B1 = reduce_sum(A, 1)
B2 = reduce_sum(A, 2)
B0, B1, B2
```

In each case we reduce the size of the tensor by summing over one of its dimensions

```
# Define features, params, and bill as constants
features = constant([[2, 24], [2, 26], [2, 57], [1, 37]])
params = constant([[1000], [150]])
bill = constant([[3913], [2682], [8617], [64400]])

# Compute billpred using features and params
billpred = matmul(features, params)

# Compute and print the error
error = bill - billpred
print(error.numpy())
```

## Advanced Operations

* Gradient - computes the slope of a function at a point
* Reshape - reshapes a tensor
* Random - populates tensor with entries drawn from a probability distribution

Finding the optimum 
* In many problems, we want to find the optimum of a function
  * Min - lowest value of a loss function
  * Max - highest value of objective function
* Can do this with `gradient()` operation
  * Optimum: find a point where gradient = 0
  * Min - change in gradient > 0
  * Max - change in gradient < 0

```
import tensorflow as tf

# Define x
x = tf.Variable(-1.0)
x
```

```
# Define y within instance of GradientTape
with tf.GradientTape() as tape:
  tape.watch(x)
  y = tf.multiply(x,x)
y
```

```
# Evaluate the gradient of y at x = -1
g = tape.gradient(y,x)
g.numpy()
```

```
def compute_gradient(x0):
  	# Define x as a variable with an initial value of x0
	x = tf.Variable(x0)
	with tf.GradientTape() as tape:
		tape.watch(x)
        # Define y using the multiply operation
		y = multiply(x,x)
    # Return the gradient of y with respect to x
	return tape.gradient(y, x).numpy()

# Compute and print gradients at x = -1, 1, and 0
print(compute_gradient(-1.0))
print(compute_gradient(1.0))
print(compute_gradient(0.0))
```
