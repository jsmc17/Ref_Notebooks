Equalizing

```
image = plt.imread('moon.jpg')
# Flatten the image into 1 dimension: pixels
pixels = image.flatten()

# Generate a cumulative histogram
cdf, bins, patches = plt.hist(pixels, bins=256, range=(0,256), density=True, cumulative=True)
new_pixels = np.interp(pixels, bins[:-1], cdf*255)

# Reshape new_pixels as a 2-D array: new_image
new_image = new_pixels.reshape(image.shape)

# Display the new image with 'gray' color map
plt.subplot(2,1,1)
plt.title('Equalized image')
plt.axis('off')
plt.imshow(new_image, cmap='gray')

# Generate a histogram of the new pixels
plt.subplot(2,1,2)
pdf = plt.hist(new_pixels, bins=64, range=(0,256), density=False,
               color='red', alpha=0.4)
plt.grid('off')

# Use plt.twinx() to overlay the CDF in the bottom subplot
plt.twinx()
plt.xlim((0,256))
plt.grid('off')

# Add title
plt.title('PDF & CDF (equalized image)')

# Generate a cumulative histogram of the new pixels
cdf = plt.hist(new_pixels, bins=64, range=(0,256),
               cumulative=True, density=True,
               color='blue', alpha=0.4)
plt.show()
```

## Statistical Plots with Seaborn

### Visualizing Regressions

.lmplot() - linear model plot

```
tips = sns.load_dataset('tips')
tips.head()
```

```
sns.lmplot(x = 'total_bill', y = 'tip', data = tips);
```

Breaking out by categorical variables

```
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', palette = 'Set1');
```

```
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex');
```

Residual plots

```
sns.residplot(x = 'total_bill', y = 'tip', data = tips, color = 'indianred');
```

### Visualizing Univariate Distributions (strip, swarm, violin)

```
sns.stripplot(y = 'tip', data = tips, jitter = True, x = 'day')
```

```
sns.swarmplot(y = 'tip', data = tips, x = 'day')
```

Changing orientation

```
sns.swarmplot(x = 'tip', data = tips, y = 'day', hue = 'sex', orient = 'h')
#switch x and y variables
```

Violinplot


```
sns.violinplot( data = tips, x = 'day', y = 'tip')
```

Combining Plots:


violinplot + stripplot

```
sns.violinplot( data = tips, x = 'day', y = 'tip', inner = None, color = 'lightgray');
sns.stripplot( data = tips, x = 'day', y = 'tip', jitter = True, size = 4);
```

### Visualizing multivariate distributions (jointplots, pairplots, heatmaps)

* Bivariate -> two variables
* Multivariate -> multiple variables
* Jointplots, pairplots, heatmaps

```
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
```

```
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')
```

Pairplot - all possible joint plots for each pair of variables

```
sns.pairplot(data = tips)
plt.show()
```

```
sns.pairplot(data = tips, hue = 'sex')
plt.show()
```

Heatmap
