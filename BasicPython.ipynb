## d-Try-Except 

The `try` block lets you test a block of code for errors.

The `except` block lets you handle the error.

The `finally` block lets you execute code, regardless of the result of the try- and except blocks.

**Exception Handling**

When an error occurs, or exception as we call it, Python will normally stop and generate an error message.

These exceptions can be handled using the `try` statement:

````python
try:
  print(float('hello'))
except:
  print("An exception occurred")
````

Can specify a specific type of error

````python
try:
  print(float(qrz)) # var not defined 
except ValueError:
  print("Must pass numerical argument")
except:
  print("Something else went wrong")
````

**Finally**

The finally block, if specified, will be executed regardless if the try block raises an error or not.

````python
try:
  print(float('hello'))
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")
````

## Flexible arguments (*args and **kwargs)

variable-length arguments (varargs):<br>
-arguments that can take an unspecified amount of input. When these are used, the programmer does not need to wrap the data in a list or an alternative sequence.

'*args

````python
def add_all(*args):
  "summ all values in *args together"
  sum_all = 0

  for num in args:
    sum_all += num

  return sum_all
````

````python
add_all(4,3,8,6.5)
````

You may be thinking that the solution with *args is very similar to the list solution. That's because *args is internally a Tuple, which is an iterable sequence similar to lists. If you'd like to verify it's type, you can enter the code in your Python interpreter:



````python
tup = 1,2,3,5,7
tup[3]
````

> Note: args is just a name, you can name that vararg anything as long as it is preceded by a single asterisk (*). It's best practice to keep naming it args to make it immediately recognizable.

'** kwargs

What makes `**kwargs` different is that it allows you to pass a variable number of keyword arguments to functions. Recall from the previous video that, within the function definition, `kwargs` is a dictionary.

````python
def print_all(**kwargs):
  """print out all key-value pairs in **kwargs."""

  for key, value in kwargs.items():
    print(key + ": " + value)
````

````python
print_all(name = 'dumbledore', job = 'headmaster')
````

Since `**kwargs` is a dictionary, you can iterate over them like any other using the `.items()` method:

````python
dikt= {'hello':1, 'world':2, 'maria':4}
dikt.items()
````

**Combining Varargs and Keyword Arguments** <br>
Quite often we want to use both `*args` and `**kwargs together`, especially when writing Python libraries or reusable code. Lucky for us, `*args` and `**kwargs` play nicely together, and we can use them in the following way:

````python
def combined_varargs(*args, **kwargs):
    print(args)
    print(kwargs)

combined_varargs(1, 2, 3, a="hi")
````

When mixing the positional and named arguments, positional arguments must come **before** named arguments. Furthermore, arguments of a fixed length come before arguments with variable length. Therefore, we get an order like this:

1. Known positional arguments
2. *args
3. Known named arguments
4. **kwargs

````python
def super_function(num1, num2, *args, callback=None, messages=[], **kwargs):
    print(str(num1 + num2))
    print(args)
    print(callback)
    if num1 % 2 == 0:
      messages.append('even')
    else:
      messages.append('odd')
    print(messages)
    print(kwargs)

super_function(1,2,10,100,kw1 = 5, kw2 = 10)
````

## Context Managers

A context manager: 
* sets up a context
* runs code
* removes the context

````python
with open('sample_data/README.md') as my_file:
  text = my_file.read()
  length = len(text)

print('The file is {} characters long'.format(length))
````

`open` does three things:
* Sets up a context by opening file
* Lets you run any code you want on file
* Removes the context by closing the file

>Using a context manager:

with < context-manager >(< args >) as < variable name >
<br>    '# run code here
<br>    '# this code is running 'inside the context'

**Writing context managers**




Two ways of writing:
* Class-based
* Function-based

Course focuses on function-based

How to create a context manager:
1. Define a function
2. (optitonal) add any set up code your context needs
3. use the 'yield' keyword (indicated special kind of function)
4. (optional) add any teardown code your context needs
5. Add the `@contextlib.contextmanager` decorator

````python
import contextlib
@contextlib.contextmanager
def my_context():
  print('hello')
  yield 42                   # yield keyword indicates value will be returned, but rest of function will be finished at some point in future
  print('goodbye')
````

````python
with my_context() as foo:
  print('foo is {}'.format(foo))
````

Timer context manager:

````python
import time
# Add a decorator that will make timer() a context manager
@contextlib.contextmanager
def timer():
  """Time the execution of a context block.

  Yields:
    None
  """
  start = time.time()
  # Send control back to the context block
  yield start
  end = time.time()
  print('Elapsed: {:.2f}s'.format(end - start))

with timer():
  print('This should take approximately 0.25 seconds')
  time.sleep(0.25)
````

**Example decorators**

> Timer wrapper:

````python
import time

def timer(func):
  """A decorator that prints how long a function took to run:

  Args: 
    func (callable): The function bein decorated.
  
  Returns:
    callable: the decorated function.
    """
  def wrapper(*args, **kwargs):
    t_start = time.time() # get current time
    result = func(*args, **kwargs) # call the decorated function and store result
    t_total = time.time() - t_start
    print(f'{func.__name__} took {t_total}s')
    return result
  return wrapper
````

````python
@timer
def sleep_n_seconds(n):
  time.sleep(n)
sleep_n_seconds(5)
````

> Not during the night:

````python
from datetime import datetime

def not_during_the_night(func):
    def wrapper():
        if 7 <= datetime.now().hour < 22:
            func()
        else:
            pass  # Hush, the neighbors are asleep
    return wrapper

@not_during_the_night
def say_whee():
    print("Whee!")

````

````python
say_whee()
#If you try to call say_whee() after bedtime (hour 22), nothing will happen:
````

> Print type of returned

````python
def print_return_type(func):
  # Define wrapper(), the decorated function
  def wrapper(*args, **kwargs):
    # Call the function being decorated
    result = func(*args, **kwargs)
    print('{}() returned type {}'.format(
      func.__name__, type(result)
    ))
    return result
  # Return the decorated function
  return wrapper
  
@print_return_type
def foo(value):
  return value
  
print(foo(42))
print(foo([1, 2, 3]))
print(foo({'a': 42}))
````

````python
def counter(func):
  def wrapper(*args, **kwargs):
    wrapper.count += 1
    # Call the function being decorated and return the result
    return func()
  wrapper.count = 0
  # Return the new decorated function
  return wrapper

# Decorate foo() with the counter() decorator
@counter
def foo():
  print('calling foo()')
  
foo()
foo()
foo()

print('foo() was called {} times.'.format(foo.count))
````

Python provides `wraps` function
* Decorator used when defining a decorator
  * Modifies wrapper's metadata to look like the function being decorated 

````python
from functools import wraps 
import time

def timer(func):
  """A decorator that prints how long a function took to run."""

  @wraps(func)
  def wrapper(*args, **kwargs):
    t_start = time.time()

    result = func(*args, **kwargs)

    t_total = time.time() - t_start
    print('{} took {}s'.format(func.__name__, t_total))
    return result
  return wrapper

@timer
def sleep_n_seconds(n=10):
  """Pause processing for n seconds

  Args:
    n (int): The number of seconds to pause for
  """
  time.sleep(n)

sleep_n_seconds()
````

**Decorators that take arguments**

A decorator factory:

Make function that *returns* a decorator rather than function that *is* a decorator

````python
def run_n_times(n):
  """Define and return a decorator"""
  def decorator(func):
    def wrapper(*args, **kwargs):
      for i in range(n):
        func(*args, **kwargs)
    return wrapper
  return decorator

@run_n_times(3)
def print_sum(a,b):
  print(a+b)

print_sum(4,5)
````

**Customizing functionality via inheritance** 

Customizing constructors
* Can run the constructor of the parent class first with `Parent.__init__(self, args...)`
* Don't *have* to call the parent constructors - can use entirely new code, but saves time/effort

````python
class BankAccount:
  def __init__(self, balance):
    self.balance = balance
  def withdraw(self, amount):
    self.balance -= amount
  BANK = 'BOA'
#Class inherited from BankAccount
class SavingsAccount(BankAccount):
  

  # Constructor specifically for Savings Account with an additional parameter
  def __init__(self, balance, interest_rate):

    # Call the parent constructor using ClassName.__init__()
    BankAccount.__init__(self, balance)  #<--- self is a SavingsAccount, but also a BankAccount


    #Add more functionality
    self.interest_rate = interest_rate
  #New method
  def compute_interest(self, n_periods = 1):
    return self.balance * ((1 + self.interest_rate) ** n_periods -1)
````

````python
class Customer:
  def __init__(self, name, balance, id):
    self.name, self.balance = name, balance
    self.id = id
  
  # Will be called when == is used
  def __eq__(self, other):
    print('__eq__() is called')

    # Returns True if all attributes match
    return (self.id == other.id) and (self.balance == other.balance) and (self.name == other.name)


c1 = Customer("Mike Pence", 4000, 123)
c2 = Customer("Mike Pence", 4000, 123)
````

## d-Best Practices for Class Design

**Inheritance and polymorphism**

Polymorphism - Using a unified interface to operate on objects of different classes

Liskov substitution principle - Base class should be interchangeable with any of its subclasses without altering any properties of the program
* Syntactically - function signatures are compatible 
  * arguments, returned values
* Semantically - the state of the object and the program remains consistent
  * subclass method doesn't strengthen input conditions
  * subclass method doesn't weaken output condiditons
  * no additional exceptions



````python
class BankAccount:
  def __init__(self, balance):
    self.balance = balance
  def withdraw(self, amount):
    self.balance -= amount
  BANK = 'BOA'
#Class inherited from BankAccount
class SavingsAccount(BankAccount):
  

  # Constructor specifically for Savings Account with an additional parameter
  def __init__(self, balance, interest_rate):

    # Call the parent constructor using ClassName.__init__()
    BankAccount.__init__(self, balance)  #<--- self is a SavingsAccount, but also a BankAccount

````

````python
def batch_withdraw(list_of_accts, amount):
  for acct in list_of_accts:
    acct.withdraw(amount)
b, s = BankAccount(1000), SavingsAccount(1000, .1)

batch_withdraw([b,s], 100)
````

Function `batch_withdraw` doesn't need to check the object to know which `withdraw()` to call.

> Violating LSP

* Syntactic incompatibility:
  * `BankAccount.withdraw()` requires 1 parameter, but `CheckingAccount.withdraw()` requires 2
* Subclass strenghening input conditions:
  * `BankAccount.withdraw()` accepts any amount, but `CheckingAccount.withdraw()` assumesthat the amount is limited
* Subclass weakening output conditions
  * `BankAccount.withdraw()` can only leave a positive balance or cause an error, but `CheckingAccount.withdraw() can leave balance negative`

Rule of thumb - if subclass violates LSP, do not use inheritance because it is likely to cause code to behave in unexpected ways

**Managing data access: private attributes**

All class data is public, but there are ways to manage access to data
* Naming conventions
* Use `@property` to customize access
* Overriding `__getattr__()`  or `__setattr__()`

<br>

Naming Conventions:
* internal attributes - `obj._att_name`, `obj._method_name()`
  * starts with a single `_` -> "internal"
  * not a part of the public API
  * nothings stopping user from accessing, but `_` is developer's way of saying not to use it
  * as class developer: use for implementation details, helper functions
  * ex - pandas has `df._is_mixed_type` attribute
* pseudoprivate attributes - `obj.__att_name`, `obj.__method_name()`
  * starts but doesn't end with `__` -> "private"
    * leading and trailing `__` only used for built-in Python methods (`__init__()`, `__repr__()`
  * data not inherited
  * name mangling: `obj.__attr_name` is interpreted as `obj._MyClass__attr_name`
  * Used to prevent name clashes in inherited classes

**Properties**

Can control attribute access
* Check value for validity
* make attrs read-only

Ex - Pandas - 
* will get error if adding column with more values than df
* shape attribute cannot be changed at all

Use property decorator - `@property`


````python
class Employee:
  def __init__(self, name, new_salary):
    self._salary = new_salary             # use protected attribute with leading _ to store data
    

  @property                               # use @property decorator on a method whose name is exactly the name of the restricted attribute
  def salary(self):
    return self._salary                   # return the internal attribute

  @salary.setter                          # use @attr.settr on a method attr() that will be called on obj.attr = value
  def salary(self, new_salary):
    if new_salary < 0:
      raise ValueError("invalid salary")
    self._salary = new_salary             # value to assign passed as argument
````

````python
emp.salary = 60000 # <--@salary.setter
````

````python
emp.salary = -1000
````

User-facing: behave like attributes
Developer-facing: give control of access

  * create a read-only property
* add `@attr.getter`
  * use for the method that is called when the property's value is retrieved
