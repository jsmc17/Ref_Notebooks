## d-Try-Except 

The `try` block lets you test a block of code for errors.

The `except` block lets you handle the error.

The `finally` block lets you execute code, regardless of the result of the try- and except blocks.

**Exception Handling**

When an error occurs, or exception as we call it, Python will normally stop and generate an error message.

These exceptions can be handled using the `try` statement:

```python
try:
  print(float('hello'))
except:
  print("An exception occurred")
```

Can specify a specific type of error

```python
try:
  print(float(qrz)) # var not defined 
except ValueError:
  print("Must pass numerical argument")
except:
  print("Something else went wrong")
```

**Finally**

The finally block, if specified, will be executed regardless if the try block raises an error or not.

```python
try:
  print(float('hello'))
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")
```

## Flexible arguments (*args and **kwargs)

variable-length arguments (varargs):<br>
-arguments that can take an unspecified amount of input. When these are used, the programmer does not need to wrap the data in a list or an alternative sequence.

'*args

```python
def add_all(*args):
  "summ all values in *args together"
  sum_all = 0

  for num in args:
    sum_all += num

  return sum_all
```

```python
add_all(4,3,8,6.5)
```

You may be thinking that the solution with *args is very similar to the list solution. That's because *args is internally a Tuple, which is an iterable sequence similar to lists. If you'd like to verify it's type, you can enter the code in your Python interpreter:



```python
tup = 1,2,3,5,7
tup[3]
```

> Note: args is just a name, you can name that vararg anything as long as it is preceded by a single asterisk (*). It's best practice to keep naming it args to make it immediately recognizable.

'** kwargs

What makes `**kwargs` different is that it allows you to pass a variable number of keyword arguments to functions. Recall from the previous video that, within the function definition, `kwargs` is a dictionary.

```python
def print_all(**kwargs):
  """print out all key-value pairs in **kwargs."""

  for key, value in kwargs.items():
    print(key + ": " + value)
```

```python
print_all(name = 'dumbledore', job = 'headmaster')
```

Since `**kwargs` is a dictionary, you can iterate over them like any other using the `.items()` method:

```python
dikt= {'hello':1, 'world':2, 'maria':4}
dikt.items()
```

**Combining Varargs and Keyword Arguments** <br>
Quite often we want to use both `*args` and `**kwargs together`, especially when writing Python libraries or reusable code. Lucky for us, `*args` and `**kwargs` play nicely together, and we can use them in the following way:

```python
def combined_varargs(*args, **kwargs):
    print(args)
    print(kwargs)

combined_varargs(1, 2, 3, a="hi")
```

When mixing the positional and named arguments, positional arguments must come **before** named arguments. Furthermore, arguments of a fixed length come before arguments with variable length. Therefore, we get an order like this:

1. Known positional arguments
2. *args
3. Known named arguments
4. **kwargs

```python
def super_function(num1, num2, *args, callback=None, messages=[], **kwargs):
    print(str(num1 + num2))
    print(args)
    print(callback)
    if num1 % 2 == 0:
      messages.append('even')
    else:
      messages.append('odd')
    print(messages)
    print(kwargs)

super_function(1,2,10,100,kw1 = 5, kw2 = 10)
```

## Context Managers

A context manager: 
* sets up a context
* runs code
* removes the context

```python
with open('sample_data/README.md') as my_file:
  text = my_file.read()
  length = len(text)

print('The file is {} characters long'.format(length))
```

`open` does three things:
* Sets up a context by opening file
* Lets you run any code you want on file
* Removes the context by closing the file

>Using a context manager:

with < context-manager >(< args >) as < variable name >
<br>    '# run code here
<br>    '# this code is running 'inside the context'

**Writing context managers**




Two ways of writing:
* Class-based
* Function-based

Course focuses on function-based

How to create a context manager:
1. Define a function
2. (optitonal) add any set up code your context needs
3. use the 'yield' keyword (indicated special kind of function)
4. (optional) add any teardown code your context needs
5. Add the `@contextlib.contextmanager` decorator

```python
import contextlib
@contextlib.contextmanager
def my_context():
  print('hello')
  yield 42                   # yield keyword indicates value will be returned, but rest of function will be finished at some point in future
  print('goodbye')
```

```python
with my_context() as foo:
  print('foo is {}'.format(foo))
```

Timer context manager:

```python
import time
# Add a decorator that will make timer() a context manager
@contextlib.contextmanager
def timer():
  """Time the execution of a context block.

  Yields:
    None
  """
  start = time.time()
  # Send control back to the context block
  yield start
  end = time.time()
  print('Elapsed: {:.2f}s'.format(end - start))

with timer():
  print('This should take approximately 0.25 seconds')
  time.sleep(0.25)
```

**Decorators**

wrapper placed around a function that changes a function's behavior

decorators are just functions that take a function as an argument and return a modified version of that function

```python
def double_args(func):
  return func
#decorator here does nothing
```

```python
@double_args
def multiply(a,b):
  return a*b

multiply(2,2)
```

Usually helpful for decorator to return new function to return

```python
def double_args(func):
  #define a new function that we can modify
  def wrapper(a,b):
    #call the passed in function, but double each argument
    return func(a * 2,b * 2)
  #return new function
  return wrapper

new_multiply = double_args(multiply)
new_multiply(1,5)
```

Decorator syntax:

```python
@double_args
def multiply(a,b):
  return a*b

multiply(2,2)
```

**Example decorators**

```python
import time

def timer(func):
  """A decorator that prints how long a function took to run:

  Args: 
    func (callable): The function bein decorated.
  
  Returns:
    callable: the decorated function.
    """
  def wrapper(*args, **kwargs):
    t_start = time.time() # get current time
    result = func(*args, **kwargs) # call the decorated function and store result
    t_total = time.time() - t_start
    print(f'{func.__name__} took {t_total}s')
    return result
  return wrapper
```

```python
@timer
def sleep_n_seconds(n):
  time.sleep(n)
sleep_n_seconds(5)
```

As an example of this, suppose we have some data stored in a list of
dictionaries:


```python
data = [{'first':'Guido', 'last':'Van Rossum', 'YOB':1956},
 {'first':'Grace', 'last':'Hopper', 'YOB':1906},
 {'first':'Alan', 'last':'Turing', 'YOB':1912}]
```

Now suppose we want to sort this data. Python has a sorted function that does this:

```python
sorted([2,4,3,5,1,6])
```

But dictionaries are not orderable: we need a way to tell the function
how to sort our data. We can do this by specifying the key function,
a function which given an item returns the sorting key for that item:


```python
# sort alphabetically by first name
sorted(data, key=lambda item: item['first'])
```

```python
# sort by year of birth
sorted(data, key=lambda item: item['YOB'])
```
